```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("flights.csv")
df
```

```python
coach = df[["coach_price"]]
coach8 = df.coach_price[df.hours == 8].reset_index()

plt.figure(figsize=(12, 6))

ax1 = plt.subplot(1,2,1)
sns.boxplot(data=coach, x=coach.coach_price)
plt.style.use(["dark_background"])
plt.title("Coach Prices")

ax2 = plt.subplot(1,2,2)
sns.boxplot(data=coach8, x=coach8.coach_price)
plt.style.use(["dark_background"])
plt.title("8 Hour Flight Coach Prices")

plt.show()
```

```python
delay_s = df[df["delay"] <60]
delay_l = df[df["delay"] >1200]

plt.figure(figsize=(12, 6))

ax1 = plt.subplot(1,2,1)
sns.boxplot(data=delay_s, x=delay_s.delay)
plt.title("Short Delays")

ax2 = plt.subplot(1,2,2)
sns.boxplot(data=delay_l, x=delay_l.delay)
plt.title("Long Delays")

plt.show()
```

```python
first = df[["firstclass_price"]]
plt.figure(figsize=(12, 6))

plt.hist(data=coach, x="coach_price", alpha=0.4)
plt.hist(data=first, x="firstclass_price", alpha=0.4)
plt.show()
```

```python
plt.figure(figsize=(16, 8))
perc = 0.1
flight_sub = df.sample(n = int(df.shape[0]*perc))
sns.scatterplot(x = "coach_price", y = "firstclass_price", data = flight_sub, s=2)
plt.show()
```

```python
plt.figure(figsize=(12, 6))

ax1 = plt.subplot(1,3,1,)
sns.histplot(data=df, x="coach_price", hue="inflight_meal")
plt.title("Meal")

ax2 = plt.subplot(1,3,2,)
sns.histplot(data=df, x="coach_price", hue="inflight_entertainment")
plt.title("Entertainment")

ax2 = plt.subplot(1,3,3,)
sns.histplot(data=df, x="coach_price", hue="inflight_wifi")
plt.title("Wifi")

plt.subplots_adjust(wspace=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 8))

sns.scatterplot(x = "hours", y = "passengers", data = df, alpha = 0.07, s=40)
plt.show()
```
